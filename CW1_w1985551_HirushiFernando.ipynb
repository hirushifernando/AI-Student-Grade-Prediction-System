## Part A  Application area review

### How AI has been applied within this problem domain

#### Introduction to AI in Student Performance Prediction
predicting student academic performance. With the increasing availability of digital educational data such as attendance records, previous grades, study time, and engagement indicators, ML models are now capable of detecting patterns that help forecast outcomes like final grades, assignment scores, and pass/fail status. Literature consistently highlights two core priorities in this domain: developing models with high predictive accuracy and ensuring that these models remain interpretable so educators can understand and trust the outputs.

#### Traditional Machine Learning Techniques in Education
Much of the existing research relies on traditional supervised machine learning methods due to their simplicity, computational efficiency, and interpretability. Algorithms such as Decision Trees, k-Nearest Neighbors (k-NN), Nave Bayes, Support Vector Machines (SVM), and Logistic or Linear Regression are frequently used as baseline models. Comparative studies indicate that logistic regression and SVM often perform strongly in classification tasks such as predicting pass/fail outcomes, while decision trees and k-NN work effectively on datasets containing a mix of categorical and numerical features. These models are preferred in many educational studies because they perform reliably even with small to medium-sized datasets and require minimal computational resources, making them accessible for academic environments.

#### Ensemble and Boosting Models
As researchers aim to improve prediction accuracy, ensemble learning methods have become increasingly prominent in literature. Approaches such as Random Forests, XGBoost, bagging, and stacking have shown significant performance gains compared to individual classifiers. Random Forest models, for example, help reduce overfitting and capture complex, nonlinear relationships among features, resulting in more stable predictions. XGBoost is frequently highlighted for its ability to achieve high R values when predicting continuous grade outcomes, particularly when socioeconomic, behavioral, and institutional variables are incorporated. Studies comparing ensemble techniques consistently report that these models outperform single algorithms across multiple evaluation metrics, demonstrating their suitability for complex educational datasets.

#### Deep Learning Approaches for Student Performance Prediction
Deep learning has also been explored in situations where large datasets or highly nonlinear relationships exist. Models such as Artificial Neural Networks (ANNs) and more advanced architectures like Activation Ensemble Deep Neural Networks (AcEn-DNN) have been proposed to capture intricate patterns in student data. For instance, AcEn-DNN integrates multiple activation functions including ReLU, tanh, sigmoid, and swish to increase model flexibility and improve learning capability. Deep learning approaches frequently achieve high accuracy and low error rates, especially in continuous grade prediction tasks. However, these methods also introduce challenges related to data requirements, hyperparameter tuning, and computational cost. Additionally, their lack of interpretability limits their suitability in real-world educational settings where teachers and administrators must clearly understand why a student is identified as at risk.

#### Explainable AI (XAI) for Model Interpretability
Given the need for transparency in educational decision making, many studies emphasize the importance of model interpretability. Explainable AI (XAI) techniques are increasingly integrated into prediction systems to help educators understand the reasoning behind model outputs. Tools such as SHAP values, LIME, and partial dependence plots are commonly used to identify the factors that contribute most to student performance, such as quiz results, class participation, assignment scores, and engagement frequency. By providing clear explanations, XAI methods offer insights that support early intervention strategies and help educators build trust in AI-driven systems. The literature shows that even when ensemble or deep learning models are used, XAI plays a critical role in ensuring the predictions are actionable and ethically aligned with educational practices.

#### Unsupervised Learning for Identifying Student Patterns
Unsupervised learning techniques also appear frequently in literature, primarily for discovering hidden student behavior patterns or grouping students based on performance characteristics. K-Means clustering is one of the most used methods and is often evaluated using cluster validation indices such as the DaviesBouldin Index. These clustering approaches help reveal subgroups of students with similar study habits, engagement levels, or academic risk profiles. Such insights can support more personalized learning interventions, particularly in online learning environments where large volumes of behavioral data from learning management systems (LMS) or MOOC platforms are available. Although clustering does not produce direct predictions, it contributes valuable understanding of student heterogeneity within educational datasets.

#### Summary of AI Techniques Used in This Domain
Overall, the literature demonstrates a clear methodology in applying AI to student performance prediction. Research typically begins with simple and interpretable models such as logistic regression and decision trees to establish baseline results. Accuracy is then enhanced through ensemble methods like Random Forests and XGBoost, which consistently outperform traditional algorithms on more complex datasets. Deep learning models are adopted when large datasets justify their complexity, though their lack of transparency limits their direct use in educational settings. Across all techniques, interpretability and fairness remain essential considerations, with XAI tools adding an important layer of transparency to ensure that predictions can be trusted and effectively used for academic support. This body of research establishes a strong foundation for developing accurate, interpretable, and practical AI-based student performance prediction systems.

## Part B  Compare and evaluate AI techniques.

### Stating the Goal Clearly

#### Goal:
The primary goal of this project is to predict a students grade category (Fail, Pass, Merit, Credit, Distinction) using historical data such as attendance, study hours, previous grades, assignment scores, parental involvement, and engagement levels. By learning from past student records, the system can estimate the likely performance of future students based on their personal and academic details. This helps identify students who may need additional support early and provides insights for data-driven academic planning.

We select three AI techniques commonly used in this domain:
- Logistic Regression
- Naive Bayes
- XGBoost


### Comparison of the Three AI Techniques

- **Technique 1: Logistic Regression**

  - **Description:** Logistic Regression is a supervised classification algorithm that models the probability of a student belonging to a specific grade category (such as Fail, Pass, Merit, Credit, or Distinction) using a logistic (sigmoid) function based on input features.

  - **Strengths:** The model is simple, computationally efficient, easy to interpret, and fast to train, making it suitable for small to medium-sized educational datasets.

  - **Weaknesses:** Logistic Regression assumes linear relationships between input features and the output and may struggle to capture complex, non-linear patterns present in student performance data.

  - **Advantages:** It is easy to implement, provides clear decision boundaries, and allows straightforward interpretation of how each feature affects student performance.

  - **Disadvantages:** Its predictive accuracy is lower compared to more advanced ensemble models when the dataset contains complex interactions between features.

  - **Input:** Attendance percentage, study hours, previous grades, assignment scores, engagement level, and parental involvement (encoded).

  - **Output:** Probability for each grade category and the final predicted student grade.

  - **Applied to Domain:** Logistic Regression is used as a baseline prediction model to understand how individual academic and behavioral factors influence student performance.

- **Technique 2: Naive Bayes**

  - **Description:** Naive Bayes is a probabilistic supervised learning model based on Bayes Theorem that predicts student grade categories by assuming conditional independence between input features.

  - **Strengths**: The model offers very fast training and prediction, performs well with small datasets, and handles categorical features efficiently after encoding.

  - **Weaknesses:** Its strong independence assumption is often unrealistic in educational data, where many student attributes are correlated, which can reduce prediction accuracy.

  - **Advantages:** Naive Bayes is simple to implement, computationally lightweight, and effective for quick classification tasks.

  - **Disadvantages:** The models performance decreases when features such as attendance, study hours, and previous grades influence each other.

  - **Input:** Numerical and categorical student attributes, including academic results and behavioral indicators.

  - **Output:** Probability distribution over grade categories and the predicted student grade.

  - **Applied to Domain:** Naive Bayes is suitable for rapidly classifying students into performance categories and identifying students who may be at academic risk.


- **Technique 3: XGBoost**

  - **Description:** XGBoost is an advanced ensemble learning technique based on Gradient Boosted Decision Trees that builds multiple weak learners sequentially to minimize prediction error.

  - **Strengths:** The model provides high prediction accuracy, effectively captures non-linear relationships, includes built-in regularization to prevent overfitting, and offers feature importance analysis.

  - **Weaknesses:** XGBoost is more computationally expensive and less interpretable compared to simpler models such as Logistic Regression.

  - **Advantages:** It handles complex structured data efficiently and performs exceptionally well on tabular datasets such as student academic records.

  - **Disadvantages:** Model tuning and training require more computational resources and expertise.

  - **Input:** Attendance, study hours, previous grades, assignment scores, engagement level, and parental involvement.

  - **Output:** Predicted student grade category (Fail, Pass, Credit, Merit, or Distinction).

  - **Applied to Domain:** XGBoost is selected as the final implementation model because it captures complex relationships in student performance data and delivers robust and accurate predictions.


Logistic Regression and Naive Bayes provide simple, interpretable baseline models for student grade prediction. However, their performance is limited when dealing with complex and non-linear relationships between academic and behavioral factors. XGBoost outperforms both techniques due to its ensemble structure, regularization mechanisms, and ability to learn complex patterns from structured educational data. Therefore, XGBoost was selected as the most suitable AI technique for the implementation of the student performance prediction system.

#### Pre-processing Steps

Before modeling with the XGBoost algorithm, the dataset underwent the following preprocessing steps to ensure data quality, consistency, and compatibility with machine learning:
dles unscaled numeric features effectively.

#### 1. Handling Missing Values
- Numerical features such as Hours_Studied, Attendance, Sleep_Hours, Previous_Scores, and New_Exam_Score contained no missing values.
- Categorical features including Parental_Involvement, Access_to_Resources, Extracurricular_Activities, Internet_Access, and Learning_Disabilities were converted into numeric representations; therefore, missing values were implicitly handled.

#### 2. Encoding Categorical Features
- All categorical attributes were converted into numeric form.
- **Ordinal features** (e.g., Parental_Involvement, Access_to_Resources) were encoded using label encoding.
- **Nominal features** (e.g., Internet_Access, Extracurricular_Activities) can be encoded using one-hot encoding.

#### 3. Feature Engineering
- Additional interaction features were created to help the model capture complex relationships between variables:
  - **Study_Sleep** (Hours_Studied  Sleep_Hours)
  - **Study_Tutoring** (Hours_Studied  Tutoring_Sessions)
  - **Resources_Internet** (Access_to_Resources  Internet_Access)

#### 4. Feature Selection
- Irrelevant and redundant columns (such as original text-based fields) were removed.
- Only meaningful numeric features were retained for model training.

#### 5. Feature Scaling
- Feature scaling was not applied, as XGBoost is tree-based and performs effectively on unscaled numeric features.

#### 6. Creating Target Labels
- New_Exam_Score was mapped to categorical performance labels:
  - Fail, Pass, Merit, Credit, Distinction
- Corresponding recommendations were assigned for each grade category.
- Grade categories and recommendations were encoded into numeric labels for model training.

#### 7. TrainTest Split
- The dataset was split into:
  - 80% training data
  - 20% testing data
- This resulted in:
  - 'train_features', 'train_target'
  - 'test_features', 'test_target'

Dataset is now clean, fully numeric, and ready for building an XGBoost model to predict student performance and provide guidance.

#### Encoding Categorical Features

```python
# Map ordinal categories to numbers
df['Parental_Involvement'] = df['Parental_Involvement'].map({'Low': 0, 'Medium': 1, 'High': 2})
df['Access_to_Resources'] = df['Access_to_Resources'].map({'Low': 0, 'Medium': 1, 'High': 2})

# Convert Yes/No to 1/0
yes_no_cols = ['Extracurricular_Activities', 'Internet_Access', 'Learning_Disabilities']
for col in yes_no_cols:
    df[col] = df[col].map({'No': 0, 'Yes': 1})
```

#### Feature Engineering

```python
from sklearn.preprocessing import MinMaxScaler

# New interaction features
df['Study_Sleep'] = df['Hours_Studied'] * df['Sleep_Hours']
df['Study_Tutoring'] = df['Hours_Studied'] * df['Tutoring_Sessions']
df['Resources_Internet'] = df['Access_to_Resources'] * df['Internet_Access']

# Update numeric features list
numeric_features = ['Hours_Studied', 'Attendance', 'Parental_Involvement', 
                    'Access_to_Resources', 'Extracurricular_Activities', 
                    'Sleep_Hours', 'Previous_Scores', 'Internet_Access', 
                    'Tutoring_Sessions', 'Learning_Disabilities',
                    'Study_Sleep', 'Study_Tutoring', 'Resources_Internet']
```

```python
# Map exam scores to grade categories 
def grade_category(score):
    if score >= 85:
        return "Distinction"
    elif score >= 70:
        return "Merit"
    elif score >= 55:
        return "Credit"
    elif score >= 40:
        return "Pass"
    else:
        return "Fail"

df['Grade'] = df['New_Exam_Score'].apply(grade_category)
```

```python
# Map grade categories to academic recommendations
recommendations = {
    "Distinction": "Outstanding performance. You are fully prepared for University Degree programs and competitive academic or professional pathways.",
    "Credit": "Satisfactory performance. Diploma or Professional Courses are recommended. With improvement, a University Degree is achievable.",
    "Merit": "Good performance. You can pursue University Degrees, Diplomas, or Professional Courses with continued effort.",
    "Pass": "You meet the minimum requirements. Certificate, NVQ, or Foundation Courses are recommended before progressing further.",
    "Fail": "Do not be discouraged. Foundation Courses combined with structured tutoring are advised to improve performance."
}

df['Recommendation'] = df['Grade'].map(recommendations)
```

```python
# Function to predict a student's grade and provide the corresponding recommendation

def predict_student_grade(student_data, model, le, recommendations):
    """
    Predicts the grade and recommendation for a new student.
    
    Parameters:
    - student_data: dict with keys matching your feature names
    - model: trained XGBoost classifier
    - le: LabelEncoder used to encode grades
    - recommendations: dict mapping grades to messages
    
    Returns:
    - predicted_grade: str
    - recommendation: str
    """
    # Create DataFrame from input
    new_student = pd.DataFrame([student_data])
    
    # Create interaction features
    new_student['Study_Sleep'] = new_student['Hours_Studied'] * new_student['Sleep_Hours']
    new_student['Study_Tutoring'] = new_student['Hours_Studied'] * new_student['Tutoring_Sessions']
    new_student['Resources_Internet'] = new_student['Access_to_Resources'] * new_student['Internet_Access']
    
    # Ensure columns match training features
    model_features = ['Hours_Studied', 'Attendance', 'Parental_Involvement', 'Access_to_Resources',
                      'Extracurricular_Activities', 'Sleep_Hours', 'Previous_Scores',
                      'Internet_Access', 'Tutoring_Sessions', 'Learning_Disabilities',
                      'Study_Sleep', 'Study_Tutoring', 'Resources_Internet']
    
    new_student = new_student[model_features]
    
    # Predict grade
    predicted_class_num = model.predict(new_student)
    predicted_grade = le.inverse_transform(predicted_class_num)[0]
    
    # Map recommendation
    recommendation = recommendations[predicted_grade]
    
    return predicted_grade, recommendation
```

```python
# Predict and display grade and recommendation for a new student

student_data = {
    "Hours_Studied": 16,
    "Attendance": 85,
    "Parental_Involvement": 2,
    "Access_to_Resources": 2,
    "Extracurricular_Activities": 1,
    "Sleep_Hours": 7,
    "Previous_Scores": 80,
    "Internet_Access": 1,
    "Tutoring_Sessions": 1,
    "Learning_Disabilities": 0
}

pred_grade, pred_recommendation = predict_student_grade(student_data, model, le, recommendations)

print("Predicted Grade:", pred_grade)
print("Recommendation:", pred_recommendation)
```

Additionally, k-fold cross-validation can be applied, where the dataset is split into k subsets and the model is trained and tested k times, each time using a different subset as the test set. This provides a more robust estimate of model performance. Model accuracy, along with metrics such as precision, recall, F1-score, and confusion matrix analysis, are used to assess how well the classifier predicts each grade category and to identify potential misclassifications.

## Part E  Equity, Diversity, Inclusion (EDI) and Sustainability

This AI-based student grade prediction and recommendation system was developed with consideration of responsible AI practices, focusing on fairness, ethical use, accessibility, and environmental sustainability.

From an Equity, Diversity, and Inclusion (EDI) perspective, the system may inherit bias from historical academic data, particularly where factors such as attendance, access to tutoring, or learning resources reflect socio-economic differences rather than student capability. To reduce bias, sensitive personal attributes such as gender, ethnicity, or financial background were intentionally excluded from the dataset. Model performance was evaluated across all grade categories to ensure balanced predictive behavior and to avoid systematic disadvantage toward any specific group of students. However, it is acknowledged that predictive outcomes should be interpreted cautiously and used only as academic guidance rather than definitive judgments.

In terms of accessibility and inclusion, the system provides consistent and scalable academic insights that can support educators in identifying students who may require early intervention. The model operates on structured academic data and can be integrated into simple digital platforms, making it accessible to institutions with limited technological resources. Nevertheless, the system depends on reliable data collection, which may limit its effectiveness in environments with inadequate digital infrastructure.

Ethically, the system is designed to assist decision-making rather than replace human judgment. Automated grade predictions and recommendations are presented as supportive insights, ensuring that educators retain control over final academic decisions. Transparency in the systems purpose and limitations is essential to prevent over-reliance on automated outcomes and to avoid negative psychological effects on students.

From an environmental sustainability perspective, the use of XGBoost for tabular data offers an efficient balance between accuracy and computational cost. Model training and evaluation were performed offline using moderate computational resources without requiring GPU acceleration. The dataset is structured and relatively small in size, resulting in minimal data storage requirements. To further reduce environmental impact in production deployment, predictions can be generated in batches, model retraining can be scheduled periodically rather than continuously, and lightweight cloud infrastructure powered by energy-efficient data centers can be utilized.

Overall, the implementation demonstrates responsible AI development by addressing societal considerations and minimizing environmental impact, while delivering meaningful academic insights to support student performance improvement.
